```python
import pandas as pd
skills=pd.read_csv("/content/skills.csv", delimiter=',')
```

```python
skills
```

```python
skill=[]
for col in skills.columns:
    skill.append(col)
```

```python
skill
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
  
# word_tokenize accepts
# a string as an input, not a file. 
stop_words = set(stopwords.words('english'))

abbre=[]
for u in skill:
  stg=""
  for word in u.split():
    if not word in stop_words : 
          # print(word[0])
          if word != "and":
            stg += str(word[0])
  abbre.append(stg.lower())
```

```python
import fuzzyset

fz1 = fuzzyset.FuzzySet()

for f in skill:
  fz1.add(f.lower())
fz2 = fuzzyset.FuzzySet()
for f in abbre:
  fz2.add(f.lower())
```

```python
fz1.get("computer scince")
```

```python
fz1.add("artificial intelligence")
fz2.add("ai")
abbre.append("ai")
skill.append("Artificial Intelligence")
```

```python
import pickle

with open("/content/skills.pth", 'wb') as fout:
    pickle.dump((fz1,fz2,skill,abbre), fout)
```
