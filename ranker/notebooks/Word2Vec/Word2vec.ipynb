```python
with ZipFile('/content/drive/MyDrive/word2vec/resumes_corpus.zip','r') as zipObj:
  listOfFileNames = zipObj.namelist()
  for fileName in listOfFileNames:
    if fileName.endswith('.txt'):
      path='drive/MyDrive/word2vec/AllResumes/'+fileName
      zipObj.extract(fileName,path)
```

```python
import os
new_list = []
for root, dirs, files in os.walk("drive/MyDrive/word2vec/AllResumes"):
    for file in files:
        if file.endswith('.txt'):
            with open(os.path.join(root, file), 'rb') as f:
                text = f.read()
                text=gensim.utils.simple_preprocess(text)
                new_list.append(text)
```

```python
model_new = gensim.models.Word2Vec(new_list, size=150, window=10, min_count=2, workers=10)
model_new.train(new_list,total_examples=len(new_list),epochs=10)
```

```python
def similairty3(cv, keyword,model):
    c=0;
    score=0;
    for x in cv:
        try:
            score=score+model.similarity(keyword,x)
            c=c+1
        except KeyError:
            continue
    if c is not 0:
        score=score*100/c
        #print(score)
    return score
```

```python
keyword=input("Enter the keyword:")
keyword=keyword.lower()
```

```python
score=0
for key in keyword:
    score=score+similairty3(resume_1,key,model)
print(score/len(keyword))
```
